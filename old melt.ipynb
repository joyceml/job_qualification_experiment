```python
cleaned_pre_format = pd.read_csv('cleaned_conj_exp_responses.csv')
cleaned_pre_format.columns
```

```python
covariates = cleaned_pre_format[['ResponseId','Q61', 'Q30', 'Q31', 'Q32', 'Q33']]
```

```python
covariates
```

```python
covariates['Q61'] = covariates['Q61'].str.replace('All of the above', '1,2,3,4,5')
covariates['Q61'] = covariates['Q61'].str.replace('Q', '')

covariates
```

```python
#joined = df.merge(covariates, left_on = 'id', right_on = 'ResponseId')
#joined
```

```python
covariates['Q61'][0]
```

```python
temp = covariates.merge(covariates['Q61'].str.split(',', expand=True), left_index = True, right_index = True)
temp
```

```python
melted = pd.melt(temp, id_vars = ['ResponseId', 'Q30', 'Q31', 'Q32', 'Q33'], value_vars = [0,1,2,3,4])
melted
```

```python
melted = melted[melted['value'].notnull()].drop(columns=['variable']).rename(columns={'Q30':'resp_ethnicity', 'Q31':'resp_sex', 'Q32': 'resp_age', 'Q33': 'resp_education'})
melted
```

```python
melted = melted.reset_index(drop=True)
melted['value'] = melted['value'].astype(int)
melted
```

```python
melted['value'][0]
```

```python
merge_final = df.merge(melted, left_on = ['id', 'choiceNum'], right_on = ['ResponseId','value'], how = 'inner')
merge_final
```

```python
merge_final = merge_final.drop(columns= ['ResponseId', 'value'])
merge_final
```

```python
merge_final.to_csv('covariate_final.csv')
```
